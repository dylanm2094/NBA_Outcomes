```python
import pandas as pd
```

```python
team = 'https://basketball.realgm.com/nba/team-stats/2020/Averages/Team_Totals/Regular_Season'
opp = 'https://basketball.realgm.com/nba/team-stats/2020/Averages/Opponent_Totals/Regular_Season'
```

```python
tables = pd.read_html(team)
team_df = tables[12]
team_df
```

```python
tables = pd.read_html(opp)
opp_df = tables[12]
opp_df
```

```python
stats_df = pd.merge(team_df,opp_df,on=['Team'])
stats_df.head()
```

```python
stats_df.columns.tolist()
```

```python
for_reb = []
opp_reb = []

for i in stats_df.iterrows():
    reb_for = i[1]['ORB_x'] + i[1]['DRB_x'] 
    reb_ag = i[1]['ORB_y'] + i[1]['DRB_y']

    for_reb.append(reb_for)
    opp_reb.append(reb_ag)

stats_df['avg_opp_rb'] = opp_reb
stats_df['avg_for_rb'] = for_reb
stats_df       
```

```python
stats_df = stats_df.rename(columns={"FGM_x":"avg_for_fg","FG%_x":"avg_for_fgp","FG%_y":"avg_opp_fgp",
                                   "3PM_x":"avg_for_3p","3P%_x":"avg_for_3pp","3P%_y":"avg_opp_3pp",
                                   "FTM_x":"avg_for_ft","FT%_x":"avg_for_ftp","FTA_y":"avg_opp_fta",
                                   "BPG_x":"avg_for_blk","BPG_y":"avg_opp_blk",
                                   "TOV_x":"avg_for_to","TOV_y":"avg_opp_to",})

stats_df = stats_df[["Team","avg_for_fg","avg_for_fgp","avg_opp_fgp","avg_for_3p","avg_for_3pp",
                   "avg_opp_3pp","avg_for_ft","avg_for_ftp","avg_opp_fta","avg_for_rb","avg_opp_rb"
                    ,"avg_for_blk","avg_opp_blk","avg_for_to","avg_opp_to"]]
stats_df
```

```python
more_stats = 'https://www.basketball-reference.com/leagues/NBA_2020.html'
tables = pd.read_html(more_stats)
more_stats_df = tables[0]
more_stats_df
```

```python
more_stats_df['Team'] = more_stats_df['Eastern Conference'].str.split('*').str[0]
more_stats_df['Team'] = more_stats_df['Team'].str.split('(').str[0]
east_df = more_stats_df
```

```python
more_stats = 'https://www.basketball-reference.com/leagues/NBA_2020.html'
tables = pd.read_html(more_stats)
more_stats_df = tables[1]
more_stats_df
```

```python
more_stats_df['Team'] = more_stats_df['Western Conference'].str.split('*').str[0]
more_stats_df['Team'] = more_stats_df['Team'].str.split('(').str[0]
west_df = more_stats_df
```

```python
cols = [0]
east_df.drop(east_df.columns[cols],axis=1,inplace=True)
west_df.drop(west_df.columns[cols],axis=1,inplace=True)
```

```python
df = pd.concat([east_df,west_df])
df = df.reset_index()
df.drop(df.columns[cols],axis=1,inplace=True)
df
```

```python
df = df.replace({'Team': {'Atlanta Hawks\xa0':'ATL','Boston Celtics':'BOS','Brooklyn Nets\xa0':'BRK',
                          'Charlotte Hornets\xa0':'CHO','Chicago Bulls\xa0':'CHI','Cleveland Cavaliers\xa0':'CLE',
                          'Dallas Mavericks\xa0':'DAL','Denver Nuggets':'DEN','Detroit Pistons\xa0':'DET',
                          'Golden State Warriors\xa0':'GSW','Houston Rockets':'HOU','Indiana Pacers':'IND',
                          'Los Angeles Clippers':'LAC','Los Angeles Lakers':'LAL','Memphis Grizzlies\xa0':'MEM',
                          'Miami Heat':'MIA','Milwaukee Bucks':'MIL','Minnesota Timberwolves\xa0':'MIN',
                          'New Orleans Pelicans\xa0':'NOP','New York Knicks\xa0':'NYK','Oklahoma City Thunder':'OKC',
                          'Orlando Magic\xa0':'ORL','Philadelphia 76ers':'PHI','Phoenix Suns\xa0':'PHO',
                          'Portland Trail Blazers\xa0':'POR','Sacramento Kings\xa0':'SAC','San Antonio Spurs\xa0':'SAS',
                          'Toronto Raptors':'TOR','Utah Jazz':'UTA','Washington Wizards\xa0':'WAS'
                    }})
df
```

```python
net_points = []

for i in df.iterrows():
    points_for =i[1]['PS/G'] * (i[1]['W']+i[1]['L'])
    points_ag =i[1]['PA/G'] * (i[1]['W']+i[1]['L'])
    net = points_for - points_ag
    net_points.append(net)

df['net_points'] = net_points
df = df.rename(columns={"W":"season_wins"})
df       
```

```python
df = df[["Team","season_wins", "net_points"]]
```
